```python
general_df = df[df['active_in_pmtct'].eq('No')]
elligible_df = general_df[(df.art_start_date < pd.to_datetime(date.today() + relativedelta(months=-3)))]
elligible_df = elligible_df.assign(elligibility_status = 'elligible')
not_elligible = general_df[(general_df.art_start_date > pd.to_datetime(date.today() + relativedelta(months=-3)))]
not_elligible = not_elligible.assign(elligibility_status = 'not_elligible')
elligible_no_vl = elligible_df[elligible_df.last_vl_result.isnull()]
elligible_with_vl = elligible_df[~elligible_df.last_vl_result.isnull()]

below_25 = elligible_with_vl[elligible_with_vl["age"] < 25]

above_25 = elligible_with_vl[elligible_with_vl["age"] > 24]


invalid_below_25 = (below_25[(below_25.last_vl_date < pd.to_datetime(date.today() + relativedelta(months=-6)))])
valid_below_25 = (below_25[(below_25.last_vl_date >= pd.to_datetime(date.today() + relativedelta(months=-6)))])
invalid_above_25 = (above_25[(above_25.last_vl_date < pd.to_datetime(date.today() + relativedelta(months=-12)))])
valid_above_25 = (above_25[(above_25.last_vl_date >= pd.to_datetime(date.today() + relativedelta(months=-12)))])


pmtct_df = df[df['active_in_pmtct'].eq('Yes')]
not_elligible_pmtct = pmtct_df[(pmtct_df.art_start_date > pd.to_datetime(date.today() + relativedelta(months=-3)))]
not_elligible_pmtct = not_elligible_pmtct.assign(elligibility_status = 'not_elligible')
elligible_pmtct = pmtct_df[(pmtct_df.art_start_date < pd.to_datetime(date.today() + relativedelta(months=-3)))]
elligible_pmtct = elligible_pmtct.assign(elligibility_status = 'elligible')
elligible_no_vl_pmtct = elligible_pmtct[elligible_pmtct.last_vl_result.isnull()]
elligible_with_vl_pmtct = elligible_pmtct[~elligible_pmtct.last_vl_result.isnull()]
valid_pmtct = (elligible_with_vl_pmtct[(elligible_with_vl_pmtct.last_vl_date > pd.to_datetime(date.today() + relativedelta(months=-6)))])
invalid_pmtct = (elligible_with_vl_pmtct[(elligible_with_vl_pmtct.last_vl_date < pd.to_datetime(date.today() + relativedelta(months=-6)))])

full_valid_df = pd.concat([valid_above_25, valid_below_25, valid_pmtct], ignore_index=True)
full_invalid_df = pd.concat([elligible_no_vl, elligible_no_vl_pmtct, invalid_pmtct, invalid_below_25, invalid_above_25], ignore_index=True)

full_valid_df = full_valid_df.assign(validity = 'valid')
full_invalid_df = full_invalid_df.assign(validity = 'invalid')
not_elligible = not_elligible.assign(validity = 'not_elligible')


final_vl_status = pd.concat([not_elligible, not_elligible_pmtct, full_valid_df, full_invalid_df], ignore_index=True)


no_blank_vl = elligible_df[~elligible_df.last_vl_result.isnull()]
no_ldl = no_blank_vl[~(no_blank_vl.last_vl_result == 'LDL')]
no_ldl.last_vl_result = no_ldl.last_vl_result.astype('int')
high_vl = no_ldl.query('last_vl_result>=200')

due_for_repeat = high_vl[(high_vl.last_vl_date < pd.to_datetime(date.today() + relativedelta(months=-3)))].query('last_vl_result>=200')

```

```python
def elligibility_category(val):
    if val == "not_elligible":
        return 'not_elligible'
    return 'elligible'

final_vl_status = (final_vl_status.assign(elligibility_status=final_vl_status.validity.apply(elligibility_category)))

def vl_category(val):
    if val >= 0 and val <200:
        return 'suppressed'
    if val >=200:
        return 'unsuppressed'
    
pivot_linelist = final_vl_status.replace('LDL',0)
pivot_linelist.last_vl_result = pd.to_numeric(pivot_linelist.last_vl_result)

pivot_linelist = pivot_linelist.assign(vl_category=pivot_linelist.last_vl_result.apply(vl_category))

#assigning age groups
def age_grouping(val):
    if val >=0 and val <10:
        return '0-9 years'
    if val >=10 and val <=14:
        return '10-14 years'
    if val >=15 and val <=19:
        return '15-19 years'
    if val >=20 and val <=24:
        return '20-24 years'
    return '25+'
pivot_linelist = pivot_linelist.assign(age_category=pivot_linelist.age.apply(age_grouping))
```

```python
validtable = pivot_linelist[pivot_linelist.validity.eq('valid')]
suppressedtable = (pivot_linelist.query('validity=="valid"').query('vl_category=="suppressed"'))

cohortvalid = validtable.pivot_table(index='art_cohort_year',
                                      columns='art_cohort_month', aggfunc={'vl_category':'count'})

cohortsuppressed = suppressedtable.pivot_table(index='art_cohort_year',
                                               columns='art_cohort_month', aggfunc={'vl_category':'count'})

cohortsuppression = cohortsuppressed.div(cohortvalid)
```
